```python
import matplotlib.pyplot as plt
# https://pythonspot.com/matplotlib-pie-chart/
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html
# https://www.geeksforgeeks.org/how-to-set-border-for-wedges-in-matplotlib-pie-chart/
```

```python
fig=plt.figure()

# coin toss
labels = 'Heads', 'Tails'
sizes = [50, 50]
# colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
colors = ['lightcoral', 'lightskyblue']
# explode = (0.1, 0, 0, 0)  # explode 1st slice
explode = (0, 0, 0, 0)  # explode 1st slice

# Plot
# plt.pie(sizes, explode=explode, labels=labels, colors=colors,
plt.pie(sizes, labels=labels, colors=colors,
    autopct=lambda x : x/100, startangle=90,
    wedgeprops = {"edgecolor" : "black"})

plt.axis('equal')
# plt.show()
fig.savefig('coin-pie.png')
```

```python
fig2=plt.figure()

# coin toss
labels = '1', '2', '3', '4', '5', '6'
sizes = [100/6]*6
# colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
colors = ['gold','yellowgreen','lightcoral', 'lightskyblue','orange','plum']
# explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, labels=labels, colors=colors,
    autopct=lambda x : '1/6', startangle=90,
    wedgeprops = {"edgecolor" : "black"})

plt.axis('equal')
# plt.show()
fig2.savefig('die-pie.png')
```
